```python
import matplotlib.pyplot as plt
import numpy as np

def sigmoid(z):
    return 1.0 / (1.0 + np.exp(-z))
```

```python
z = np.arange(-7, 7, 0.1)
sigma_z = sigmoid(z)

plt.plot(z, sigma_z)
plt.axvline(0.0, color = "k") # 수직선 그리는 코드
plt.ylim(-0.1, 1.1)
plt.xlabel('z')
plt.ylabel('$\sigma (z)$')

plt.yticks([0.0, 0.5, 1.0])
ax = plt.gca()
ax.yaxis.grid(True)
plt.show()
```

$$\sigma(z) = P(Y = 1 | w, x, b)$$

$$ \sigma(z) > 0.5 : \hat{y} = 1 $$

# Loss function of logistic regrassion

$$ 
Y_i | x ; w, b \sim Bernoulli(\sigma(z)) \\ \ \\
L(w, b | (x, y)) = \prod_{i = 1}^nP(y_i | x_i, w, b) 
= \prod_{i=1}^n [\sigma(z_i)^{y_i}][1-\sigma(z_i)^{1-y_i}] \\ \ \\

l(w, b | x, y) = \sum_{i = 1}^n [y_i log\sigma({z_i}) + (1-y_i)log(1-\sigma(z_i))]
$$

```python
def loss_1(z):
    return -np.log(sigmoid(z))

def loss_0(z):
    return -np.log(1-sigmoid(z))

z = np.arange(-10, 10, 0.1)
sigma_z = sigmoid(z)

c1 = [loss_1(x) for x in z]
plt.plot(sigma_z, c1, label = 'L(w, b) if y = 1')

c0 = [loss_0(x) for x in z]
plt.plot(sigma_z, c0, linestyle = '--',  label = 'L(w, b) if y = 0')
plt.ylim(0.0, 5.1)
plt.xlim([0, 1])
plt.show()
```

$$Loss(w, b) = \frac{1}{n}\sum_{i=1}^n [-y_i(log(\sigma(z_i))) - (1-y_i)(log(1-\sigma(z_i)))]
